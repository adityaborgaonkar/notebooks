# NewsAPI

## Basics
### Generate your API key >> https://newsapi.org/

```python
api_key = input("Enter Your NewsAPI key :: ")
```

```python
import requests
```

## 101

```python
url = ('http://newsapi.org/v2/top-headlines?'
       'country=us&'
       'apiKey='+api_key)

response = requests.get(url)

print(response.json())
```

## Get news by keyword

```python
url = ('http://newsapi.org/v2/everything?'
       'q=Apple&'
       'from=2020-09-18&'
       'sortBy=popularity&'
       'apiKey='+api_key)

response = requests.get(url)

print(response.json())
```
